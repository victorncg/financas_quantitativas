```python
import pandas as pd
import yfinance as yf

# Carregando dados de preços de ações
df = yf.download('PETR4.SA', start = '2021-01-01')

# Calculando o retorno diário
df['Retorno'] = df['Close'].pct_change()
```

```python
df['Retorno'].plot(figsize = (12,7));
```
