```python
ibov = [i + '.SA' for i in ad.index_composition('ibov', mode  = 'list')]
```

```python
ibov
```

```python
ibovdf = yf.download(ibov, start = '2024-01-01')['Adj Close']
```

```python
ibovdf_norm = ibovdf/ibovdf.iloc[0]
```

```python
list(ibovdf_norm.iloc[-1].sort_values().tail(10).index)
```

```python
ibovdf_filter = ibovdf[list(ibovdf_norm.iloc[-1].sort_values().tail(10).index)]
```

```python
ibovdf_filter
```

```python
retornos = ibovdf_filter.pct_change()
```

```python
retornos = retornos.iloc[1:]
```

```python
import seaborn as sns
```

```python
sns.pairplot(retornos, diag_kind ='hist')
```
